```python
# import librarias
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#line plot 
x=[1,2,3,4,5]
y=[10,45,65,64,56]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("data1")
plt.ylabel("data2")
plt.show()
```

```python
#Bar plot
import pandas as pd
data=pd.DataFrame({
    "Class":['A','B','C','D','E'],
    "Students":[99,78,55,57,58]
})    
sns.barplot(x='Class',y='Students',data=data)
plt.title("Bar plot")
plt.xlabel("Class")
plt.ylabel("Students")
plt.show()
```

```python
#count plot
sns.countplot(x='Class',data=data)
plt.title("count plot")
plt.show()
```

```python
#Histogram
import numpy as np
data=np.random.randn(100)
plt.hist(data,bins=10)
plt.title("Histogram")
plt.show()
```

```python
#KDE plot
sns.kdeplot(data,shade=True)
plt.title("KDEplot")
plt.show()
```

```python
#histogram with KDE
sns.histplot(data, bins=12, kde=True) 
plt.title("Histogram with KDE") 
plt.show() 
```

```python
# box plot
sns.boxplot(y=data)
plt.title("Box plot")
plt.show()
```

```python
#Violin plot
sns.violinplot(y=data)
plt.title("violin plot")
plt.show()
```

```python
#Scatter plot
#using matplot
x=[1,2,3,4,5]
y=[10,7,5,4,6]
plt.scatter(x,y)
plt.title("Scatter plot")
plt.show()
```

```python
#using seaborn
sns.scatterplot(x=x,y=y)
plt.title("Scatter plot")
plt.show()
```

```python
#pair 
#loading data
df=sns.load_dataset('iris')
df
#pair plot
sns.pairplot(df,hue='species')
plt.title("pair plot")
plt.show()
```

```python
#Heatmap
#df_numeric=df.drop(columns='species')
df_numeric=df.drop('species',axis=1)
correlation_matrix = df_numeric.corr() 
correlation_matrix 
#Heatmap
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

```python
#Piechart
labels=['A','B','C']
sizes=[30,40,30]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=0)
plt.title("pie chart")
plt.show()
```

```python
#Swarm plot
sns.swarmplot(x='species',y='sepal_length',data=df)
plt.title("Swarm plot")
plt.show()
```

```python
#Strip plot
sns.stripplot(x='species',y='sepal_length',data=df,jitter=True)
plt.title("Strip plot")
plt.show()
```

```python
#join plot
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='scatter')
plt.title("Swarm plot")
plt.show()
```

```python
#Rug plot
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='scatter')
plt.title("Swarm plot")
plt.show()
```
