```python
import pandas as pd
```

```python
# sample data
data={
    'CustomerID':['C001','C002','C003','C004','C005','C006','C007','C008'],
    'Education_Level':['HighSchool','Bachlors','Masters','PhD','Bachlors','Masters','PhD','HighSchool'],
    'Product_Qulity':['Low','Medium','High','Low','High','Medium','Low','Medium'],
    'Satisfaction_level':['Poor','Averge','Exellent','Good','Poor','Averge','Good','Exellent']
}
df=pd.DataFrame(data)
```

```python
df
```

```python
#Define the order for each ordinal feature

Education_order=['HighSchool','Bachlors','Masters','PhD']
Product_Qulity_order=['Low','Medium','High']
Satisfaction_level=['Poor','Averge','Good','Exellent']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder=OrdinalEncoder(categories=[Education_order,Product_Qulity_order,Satisfaction_order])
encoder
```

```python
# Appliy encoding (skip CustomerId as it is identidfier

df[['Education_Level','Product_Qulity','Satisfaction_level']]= encoder.fit_transform(
df[['Education_Level','Product_Qulity','Satisfaction_level']])
```
