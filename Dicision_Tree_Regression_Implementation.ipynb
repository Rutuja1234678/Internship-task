```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing and reading dataset

ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
# spliting dataset into Independent Ana Dependent variables
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
# Fitting Dicision Tree Regression Model

from sklearn.tree import DecisionTreeRegressor,plot_tree
reg=DecisionTreeRegressor(random_state=0)
reg.fit(x,y)
```

```python
#predict for myinput
myinput=pd.DataFrame(data=[[4]],columns=['Level'])
reg.predict(myinput)
```

```python
# visulize Decision tree
plt.figure(figsize=(70,70)) # set size of figure
plot_tree(reg,feature_names=['Level'],filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```
